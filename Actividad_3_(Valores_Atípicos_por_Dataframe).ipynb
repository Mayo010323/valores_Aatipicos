```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/Gastos y costos 20-23.xlsx", sheet_name = '2023', header = 3 )
df
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME

```python
data1 =df.fillna(method="bfill")
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME

```python

#Método aplicando desviación estandar. Encuentro los valores extremos
limite_superior = df.mean() + 3*df.std()
limite_inferior = df.mean() - 3*df.std()
print("Límite superior permitido" , limite_superior)
print("\nLímite inferior permitido" , limite_inferior)
```

```python
df2 = df[(df <= limite_superior) & (df >=limite_inferior)]
df2
```

```python
#Eliminamos columnas innecesarias
df3 = df2.drop(["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO","Descripción","Tipo", "Status", "Poliza","Otros","MP"], axis=1)
df3
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#sustuir con valores de media
df_clean = df3.copy()
df_clean = df_clean.fillna(round(df3.mean(),1))
df_clean
```

```python
df_clean.isnull().sum()
```

PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME

```python
y=df_clean

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data3_iqr= df_clean[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
nulos = data3_iqr.isnull().sum()
nulos
```

```python
#remplazamos valores atípicos
df5_iqr = data3_iqr.copy()
df5_iqr = data3_iqr.fillna(round(data3_iqr.mean(),1))
df5_iqr
```

```python
#unimos columnas string eliminadas
columnas = [ "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"]
indice= df[columnas]
datos_limpios = pd.concat([indice, df5_iqr],axis=1)
datos_limpios
```

```python
fig = plt.figure(figsize =(15, 8))
df5_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```
